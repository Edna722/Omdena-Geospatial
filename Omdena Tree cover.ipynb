```python
import rasterio

# Define the paths to your TIFF files
tif_files = [
    r"C:\Users\wanji\Desktop\Raster Files\Rasters\Rasters\Machakos_2017.tif",
    r"C:\Users\wanji\Desktop\Raster Files\Rasters\Rasters\Machakos_2019.tif",
    r"C:\Users\wanji\Desktop\Raster Files\Rasters\Rasters\Machakos_2021.tif",
    r"C:\Users\wanji\Desktop\Raster Files\Rasters\Rasters\Machakos_2023.tif"
]

# Function to read and display information about a raster file
def read_raster(file_path):
    with rasterio.open(file_path) as src:
        raster_data = src.read()  # Read the raster data
        metadata = src.profile  # Get the metadata
        crs = src.crs  # Coordinate Reference System
        bounds = src.bounds  # Spatial extent
        height, width = src.height, src.width  # Dimensions
        xres, yres = src.res  # Spatial resolution
        
        print(f"File: {file_path}")
        print(f"CRS: {crs}")
        print(f"Bounds: {bounds}")
        print(f"Dimensions: {width} x {height}")
        print(f"Resolution: {xres}, {yres}")
        print(f"Metadata: {metadata}")
        print("--------------------------------------------------")

# Read and display information for each TIFF file
for tif_file in tif_files:
    read_raster(tif_file)
```

## Visual presentation of the tree cover for the years 2017-2023

```python
# Function to load and display basic information about the shapefile
def load_and_explore(shapefile_path):
    data = gpd.read_file(shapefile_path)
    print(data.head())
    print(data.info())
    print(data.describe())
    return data

# Paths to the shapefiles
shapefile_paths = {
    "2017": r"C:\Users\wanji\Desktop\Omdena 2\Machakos_Trees (1)\Machakos_Trees\Machakos_trees_2017.shp",
    "2019": r"C:\Users\wanji\Desktop\Omdena 2\Machakos_Trees (1)\Machakos_Trees\Machakos_trees_2019.shp",
    "2021": r"C:\Users\wanji\Desktop\Omdena 2\Machakos_Trees (1)\Machakos_Trees\Machakos_trees_2021.shp",
    "2023": r"C:\Users\wanji\Desktop\Omdena 2\Machakos_Trees (1)\Machakos_Trees\Machakos_trees_2023.shp",
}

# Load and explore data for each year
data_dict = {year: load_and_explore(path) for year, path in shapefile_paths.items()}

# Plot the tree cover for each year
fig, axes = plt.subplots(2, 2, figsize=(15, 15))
axes = axes.flatten()
for i, (year, data) in enumerate(data_dict.items()):
    data.plot(ax=axes[i], color="green")
    axes[i].set_title(f"Tree Cover {year}")
plt.tight_layout()
plt.show()
```

## Temporal Analysis 
* Analyzing temporal trends in tree cover over the years by plotting the total tree cover area against time providing insights into long-term changes in tree cover.
* Visualize the annual rate of change in tree cover area to identify periods of rapid increase or decrease in tree cover.

```python
# Temporal Analysis
# Plot total tree cover area against time
years = ["2017", "2019", "2021", "2023"]
total_tree_cover_area = [data_dict[year]['geometry'].area.sum() for year in years]

plt.plot(years, total_tree_cover_area, marker='o')
plt.title("Temporal Trend of Tree Cover Area")
plt.xlabel("Year")
plt.ylabel("Total Tree Cover Area (square meters)")
plt.grid(True)
plt.show()

```

## Spatial Analysis: Tree cover

```python
# Function to calculate and return the area of tree cover
def calculate_tree_cover_area(data):
    data = data.to_crs(epsg=3395)  # Project to a coordinate system with meter-based units
    data["area"] = data.geometry.area
    total_area = data["area"].sum()
    return total_area

# Calculate tree cover area for each year
tree_cover_areas = {year: calculate_tree_cover_area(data) for year, data in data_dict.items()}

# Print the tree cover areas
for year, area in tree_cover_areas.items():
    print(f"Total tree cover area in {year}: {area:.2f} square meters")

# Calculate differences in tree cover area between the years
print("\nDifferences in tree cover area:")
for year1, year2 in zip(["2017", "2019", "2021"], ["2019", "2021", "2023"]):
    difference = tree_cover_areas[year2] - tree_cover_areas[year1]
    print(f"Change from {year1} to {year2}: {difference:.2f} square meters")
```

## Plot for the Spatial analysis 

```python
import matplotlib.pyplot as plt

# Calculate differences in tree cover area between the years
differences = []
for year1, year2 in zip(["2017", "2019", "2021"], ["2019", "2021", "2023"]):
    difference = tree_cover_areas[year2] - tree_cover_areas[year1]
    differences.append(difference)

# Plot the change in tree cover detection
years = ["2017-2019", "2019-2021", "2021-2023"]
plt.bar(years, differences, color=['green' if diff >= 0 else 'red' for diff in differences])
plt.title("Change in Tree Cover Detection (2017-2023)")
plt.xlabel("Year Range")
plt.ylabel("Change in Tree Cover Area (square meters)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Tree Cover Density 

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Function to calculate tree cover density
def calculate_tree_cover_density(data):
    data = data.to_crs(epsg=3395)  # Project to a coordinate system with meter-based units
    data["area"] = data.geometry.area
    total_area = data["area"].sum()
    total_count = len(data)
    density = total_count / total_area
    return density

# Calculate tree cover density
tree_cover_density = calculate_tree_cover_density(merged_data)
print("Tree cover density:", tree_cover_density)

# Plot the density as a choropleth map
fig, ax = plt.subplots(figsize=(10, 10))
merged_data.plot(column=None, ax=ax, alpha=0.5, edgecolor="k")
ax.set_title("Tree Cover Density")

# Add colorbar
sm = plt.cm.ScalarMappable(cmap="viridis", norm=plt.Normalize(vmin=0, vmax=tree_cover_density))
sm._A = []
cbar = fig.colorbar(sm)
cbar.set_label("Tree Cover Density")

plt.show()
```
