```python
#loading libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import osmnx as ox
import geopandas as gpd
import rasterio
```

```python
#loading Open Street Maps
place = 'Machakos county , Kenya'

# Retrieving the street network data for the defined location
graph = ox.graph_from_place(place, network_type='all')

# Plotting the street network
ox.plot_graph(graph)
```

```python
#Reading Data
omdena = gpd.read_file (r"C:\Users\wanji\Desktop\Omdena tree")
```

```python
 File paths
file_paths = [
    r"C:\Users\wanji\Desktop\Omdena tree\Machakos_2015.tif",
    r"C:\Users\wanji\Desktop\Omdena tree\Machakos_2020.tif",
    r"C:\Users\wanji\Desktop\Omdena tree\Machakos_2023.tif"
]

# Reading the raster files
try:
    with rasterio.open(file_paths[0]) as src_2015, \
         rasterio.open(file_paths[1]) as src_2020, \
         rasterio.open(file_paths[2]) as src_2023:

        # Read the raster data
        raster_2015 = src_2015.read()
        raster_2020 = src_2020.read()
        raster_2023 = src_2023.read()

        # Get the metadata
        metadata_2015 = src_2015.profile
        metadata_2020 = src_2020.profile
        metadata_2023 = src_2023.profile

        # Get the coordinate reference system
        crs = src_2015.crs

        # Get the spatial extent
        bounds = src_2015.bounds

        # Get the raster dimensions
        height_2015, width_2015 = src_2015.height, src_2015.width
        height_2020, width_2020 = src_2020.height, src_2020.width
        height_2023, width_2023 = src_2023.height, src_2023.width

        # Get the raster's spatial resolution
        xres_2015, yres_2015 = src_2015.res
        xres_2020, yres_2020 = src_2020.res
        xres_2023, yres_2023 = src_2023.res

    # Now you can use the variables for further processing
    print("Raster files loaded successfully.")

except Exception as e:
    print("Error reading raster files:", e)
```

```python
# Plot raster
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].imshow(raster_2015[0], cmap='gray')
axs[0].set_title('Raster 2015')

axs[1].imshow(raster_2020[0], cmap='gray')
axs[1].set_title('Raster 2020')

axs[2].imshow(raster_2023[0], cmap='gray')
axs[2].set_title('Raster 2023')

# Plot vector
omdena.plot(ax=axs[0], color='red', markersize=5)
omdena.plot(ax=axs[1], color='red', markersize=5)
omdena.plot(ax=axs[2], color='red', markersize=5)

plt.tight_layout()
plt.show()
```

```python
#Visulizing Machakos Spatial Distribution
omdena.plot(column='Shape_Leng', cmap='viridis', legend=True)
plt.title("Machakos Spatial Distribution")
plt.show()
```

```python
#Machakos Aggregated Data
agg_data = omdena.groupby('geometry')['Shape_Leng'].sum()
agg_data
```

```python
#Machakos Spatial Queries
from shapely.geometry import box

# Defining coordinates for the bounding box
min_x, min_y, max_x, max_y = 36.8, -1.8, 37.2, -1.3

# Creating the bounding box
bbox = gpd.GeoSeries([box(min_x, min_y, max_x, max_y)], crs=omdena.crs)

sel_features = omdena[omdena.intersects(bbox.unary_union)]
sel_features
```

```python
# Create a correlation matrix
corr_matrix = omdena.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Group data by a categorical variable and calculate mean values
grouped_data = omdena.groupby('category')['Shape_Leng'].mean()
grouped_data.plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Mean Shape_Leng')
plt.title('Mean Shape_Leng by Category')
plt.show()
```
